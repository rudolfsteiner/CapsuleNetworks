````python
from MyCNets import Config, CapsuleNetworksModel, tweak_pose_parameters
````

````python
#def do_train(mnist, restore_checkpoint):

restore_checkpoint = True
config = Config()
caps_net_model = CapsuleNetworksModel(config)

n_iterations_per_epoch = mnist.train.num_examples // config.batch_size
n_iterations_validation = mnist.validation.num_examples // config.batch_size
best_loss_val = np.infty
checkpoint_path = "./my_capsule_network2"

init = tf.global_variables_initializer()
saver = tf.train.Saver()
with tf.Session() as sess:
    if restore_checkpoint and tf.train.checkpoint_exists(checkpoint_path):
        saver.restore(sess, checkpoint_path)
    else:
        init.run()

    for epoch in range(config.num_epochs):
        for iteration in range(1, n_iterations_per_epoch + 1):
            X_batch, y_batch = mnist.train.next_batch(config.batch_size)
            # Run the training operation and measure the loss:

            loss_train = caps_net_model.train_on_batch(sess, X_batch.reshape([-1, 28, 28, 1]), y_batch, True)
            print("\rIteration: {}/{} ({:.1f}%)  Loss: {:.5f}".format(
                      iteration, n_iterations_per_epoch,
                      iteration * 100 / n_iterations_per_epoch,
                      loss_train),
                  end="")

        # At the end of each epoch,
        # measure the validation loss and accuracy:
        loss_vals = []
        acc_vals = []
        correct_vals = []
        for iteration in range(1, n_iterations_validation + 1):
            X_batch, y_batch = mnist.validation.next_batch(config.batch_size)

            _, correct_val, loss_val, acc_val = caps_net_model.validate_on_batch(sess, X_batch.reshape([-1, 28, 28, 1]), y_batch)
            loss_vals.append(loss_val)
            acc_vals.append(acc_val)
            correct_vals.append(correct_val)
            print("\rEvaluating the model: {}/{} ({:.1f}%)".format(
                      iteration, n_iterations_validation,
                      iteration * 100 / n_iterations_validation),
                  end=" " * 10)
        loss_val = np.mean(loss_vals)
        acc_val = np.mean(acc_vals)
        print("\rEpoch: {}  Val accuracy: {:.4f}%  Loss: {:.6f}{}".format(
            epoch + 1, acc_val * 100, loss_val,
            " (improved)" if loss_val < best_loss_val else ""))

        # And save the model if it improved:
        if loss_val < best_loss_val:
            save_path = saver.save(sess, checkpoint_path)
            best_loss_val = loss_val
````

# Makes some prediction through the mnist test set:

````python
n_samples = 5

sample_images = mnist.test.images[:n_samples].reshape([-1, 28, 28, 1])

with tf.Session() as sess:
    saver.restore(sess, checkpoint_path)

    caps2_output_value, decoder_output_value, y_pred_value = caps_net_model.predict_on_batch(sess, sample_images)
````

````python
caps2_output_value.shape
````

````python
sample_images = sample_images.reshape(-1, 28, 28)
reconstructions = decoder_output_value.reshape([-1, 28, 28])

plt.figure(figsize=(n_samples * 2, 3))
for index in range(n_samples):
    plt.subplot(1, n_samples, index + 1)
    plt.imshow(sample_images[index], cmap="binary")
    plt.title("Label:" + str(mnist.test.labels[index]))
    plt.axis("off")

plt.show()

plt.figure(figsize=(n_samples * 2, 3))
for index in range(n_samples):
    plt.subplot(1, n_samples, index + 1)
    plt.title("Predicted:" + str(y_pred_value[index]))
    plt.imshow(reconstructions[index], cmap="binary")
    plt.axis("off")
    
plt.show()
````

# Interpreting the Output Vectors

Let's tweak the output vectors to see what will happen


````python
n_steps = 11

tweaked_vectors = tweak_pose_parameters(caps2_output_value, n_steps=n_steps)
tweaked_vectors_reshaped = tweaked_vectors.reshape(
    [-1, config.caps2_num_caps, config.caps2_num_dims])
````

````python
tweaked_labels = np.tile(mnist.test.labels[:n_samples], config.caps2_num_dims * n_steps)
````

````python
print(tweaked_vectors_reshaped.shape)
````

````python
with tf.Session() as sess:

    saver.restore(sess, checkpoint_path)

    tweaked_decoder_output_value = caps_net_model.predict_on_tweaked_batch(sess, 
                                                                           tweaked_vectors_reshaped, tweaked_labels)
````

````python
tweak_reconstructions = tweaked_decoder_output_value.reshape(
        [config.caps2_num_dims, n_steps, n_samples, 28, 28])
````

````python
for dim in range(16):
    print("Tweaking output dimension #{}".format(dim))
    plt.figure(figsize=(n_steps / 1.2, n_samples / 1.5))
    for row in range(n_samples):
        for col in range(n_steps):
            plt.subplot(n_samples, n_steps, row * n_steps + col + 1)
            plt.imshow(tweak_reconstructions[dim, col, row], cmap="binary")
            plt.axis("off")
    plt.show()
````
